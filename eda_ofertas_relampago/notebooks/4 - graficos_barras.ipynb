```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
load_path = '../data/processed/tratamento_0002.csv'
df_load = pd.read_csv(load_path)
```

```python
print(df_load.columns)
```

```python
df_bar1 = df_load.groupby(["VERTICAL"],as_index=False)["SOLD_QUANTITY"].count().sort_values(by=["SOLD_QUANTITY"])
```

```python
FIGSIZE = 20
plt.figure(figsize = (FIGSIZE,FIGSIZE/2))
x = df_bar1["SOLD_QUANTITY"]
y = df_bar1["VERTICAL"]
plt.barh(y, x)
```

```python
df_bar2 = df_load.groupby(["VERTICAL"],as_index=False)["SOLD_AMOUNT"].count().sort_values(by=["SOLD_AMOUNT"])
```

```python
FIGSIZE = 20
plt.figure(figsize = (FIGSIZE,FIGSIZE/2))
x = df_bar2["SOLD_AMOUNT"]
y = df_bar2["VERTICAL"]
plt.barh(y, x)
```
