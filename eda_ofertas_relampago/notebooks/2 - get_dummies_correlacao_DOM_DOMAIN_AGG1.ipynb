```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
load_path = '../data/processed/tratamento_0002.csv'
df_load = pd.read_csv(load_path)
```

```python
df_load['OFFER_START_DATE'] = pd.to_datetime(df_load['OFFER_FINISH_DTTM'])
df_load['OFFER_START_DTTM'] = pd.to_datetime(df_load['OFFER_START_DTTM'])
df_load['OFFER_FINISH_DTTM'] = pd.to_datetime(df_load['OFFER_FINISH_DTTM'])
df_load["WEEK_DAY"] = df_load['OFFER_FINISH_DTTM'].dt.dayofweek
df_load["HOUR"] = df_load['OFFER_FINISH_DTTM'].dt.hour
```

```python
df_load['DOM_DOMAIN_AGG1'] = df_load['DOM_DOMAIN_AGG1'].str.replace(' ','_')
df_load.head(5)
```

```python
df_load_dummies = pd.get_dummies(df_load, columns=["DOM_DOMAIN_AGG1"])
```

```python
df_load_dummies.head(3)
```

```python
# df_load_dummies.columns.unique()
```

```python
for col_agg_name in df_load["DOM_DOMAIN_AGG1"].unique():
    df_load_dummies[f"DOM_DOMAIN_AGG1_{col_agg_name}"] = df_load_dummies["SOLD_AMOUNT"]*df_load_dummies[f"DOM_DOMAIN_AGG1_{col_agg_name}"]
```

```python
df_load_dummies.head()
```

```python
df_load_dummies.drop(['OFFER_START_DATE', 'OFFER_START_DTTM', 'OFFER_FINISH_DTTM', 'VERTICAL', 'DOMAIN_ID'], axis=1, inplace=True)
df_load_dummies.head()
```

```python
%%time
corrmat = df_load_dummies.corr(method="spearman")
corrmat_limit = corrmat[corrmat > 0.01]
FIGSIZE = 100
f, ax = plt.subplots(figsize =(FIGSIZE*3, FIGSIZE))
sns.heatmap(corrmat_limit, ax = ax, cmap ="YlGnBu", linewidths = 0.1, annot = True, annot_kws={"size":16})
```
