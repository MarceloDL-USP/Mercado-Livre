```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
load_path = '../data/processed/tratamento_0002.csv'
df_load = pd.read_csv(load_path)
```

```python
%%time
corrmat = df_load.corr(method="spearman")
FIGSIZE = 12
f, ax = plt.subplots(figsize =(FIGSIZE*2, FIGSIZE))
sns.violinplot(x= "SOLD_AMOUNT", y="VERTICAL",
                    data=df_load,
                    cut=0, scale="width", palette="Set3")
```

```python
LIMITE = 100
df_load_limit = df_load.query(f'SOLD_AMOUNT < {LIMITE}')
```

```python
%%time
corrmat = df_load_limit.corr(method="spearman")
FIGSIZE = 20
f, ax = plt.subplots(figsize =(FIGSIZE, FIGSIZE))
sns.violinplot(x= "SOLD_AMOUNT", y="VERTICAL",
                    data=df_load_limit,
                    cut=0, scale="width", palette="Set3")
```
